```python
import torch
from diffusers import StableDiffusionPipeline

# Load the pre-trained model on CPU
pipe = StableDiffusionPipeline.from_pretrained("runwayml/stable-diffusion-v1-5", torch_dtype=torch.float32)
pipe = pipe.to("cpu")  # Use CPU since GPU is not available in free Colab

print("Model loaded successfully!")
```

```python
# Take user input for the prompt
prompt = input("Enter a prompt to generate an image: ")

# Generate the image
print("Generating image... This may take a while on CPU.")
image = pipe(prompt).images[0]

# Save the image
filename = "generated_image.png"
image.save(filename)
print(f"Image saved as '{filename}'")

# Display the image
from PIL import Image
Image.open(filename).show()
```
